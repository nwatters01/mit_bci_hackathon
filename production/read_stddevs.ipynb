```python
"""Imports."""

import matplotlib.pyplot as plt
import numpy as np
```

```python
"""Read, plot, and print stddevs."""

_DATA_PATH = 'data/stddevs_v0.csv'
data_array = np.genfromtxt(_DATA_PATH, delimiter=',', dtype=float)
print(f'data_array.shape = {data_array.shape}')

fig, ax = plt.subplots(figsize=(4, 3))
ax.plot(data_array)
ax.set_xlabel('Time (batches)')
ax.set_ylabel('Stddev')

print(np.round(np.mean(data_array, axis=0), decimals=3))
```
